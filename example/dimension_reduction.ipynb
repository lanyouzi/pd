```python
import pandas as pd
import json
import numpy as np
import os
import shutil
import matplotlib
matplotlib.use('agg')
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
from sklearn.manifold import Isomap, SpectralEmbedding   
```

```python
eval_dir = 'privacy_eval'
if not os.path.exists(eval_dir):
    os.mkdir(eval_dir)
```

```python
def pca_analysis(feature, label, path):
    plt.figure(figsize=(15, 8))
    pca = PCA(n_components=40)
    pca.fit(feature)
    new_X_pca = pca.transform(feature)
    # print(np.cumsum(pca.explained_variance_ratio_))
    plt.scatter(new_X_pca[:,0], new_X_pca[:,1], c=label)
    plt.title('PCA')
    plt.savefig(os.path.join(path, 'pca.png'))

def mds_analysis(feature, label, path):
    plt.figure(figsize=(15, 8))
    mds = MDS(n_components=2, metric=True)
    new_X_mds = mds.fit_transform(feature)
    print(np.cumsum(mds.explained_variance_ratio_))
    plt.scatter(new_X_mds[:,0], new_X_mds[:,1], c=label)
    plt.title('MDS')
    plt.savefig(os.path.join(path, 'mds.png'))
    plt.close()

def isomap_analysis(feature, label, path):
    plt.figure(figsize=(15, 8))
    n_neighbors = 3
    isomap = Isomap(n_components=2, n_neighbors=n_neighbors)
    new_X_isomap = isomap.fit_transform(feature)
    plt.scatter(new_X_isomap[:,0], new_X_isomap[:,1], c=label)
    plt.title('ISOMAP')
    plt.savefig(os.path.join(path, 'isomap.png'))plt.close()

def se_analysis(feature, label, path):
    plt.figure(figsize=(15, 8))
    spectral_embedding = SpectralEmbedding(n_components=2)
    new_X_se = spectral_embedding.fit_transform(feature)
    plt.scatter(new_X_se[:,0], new_X_se[:,1], c=label)
    plt.title('Spectral Embedding')
    plt.savefig(os.path.join(path, 'se.png'))
```

```python
code_path = '/mnt/DataCenter/maruokai/datasets/pd/code/similarity_learning_4/hash_codes'
health_codes = [str(item.split('.')[0]) for item in os.listdir(code_path)]
# print(health_codes)
df = pd.read_csv('/mnt/DataCenter/maruokai/datasets/pd/disease_history_analysis.csv')
df = df[df['healthCode'].isin(health_codes)]
df.set_index(['age', 'gender'], inplace=True)

    
```

```python
history_dir = os.path.join(eval_dir, 'history_eval')
if not os.path.exists(history_dir):
    os.mkdir(history_dir)
for idx in df.index.unique():
    history_df = df.loc[idx]
    if 'No Head Injury' in history_df['health-history'].values and 'Head Injury' in history_df['health-history'].values:
        cp = os.path.join(history_dir, str(idx))
        if not os.path.exists(cp):
            os.mkdir(cp)
        print(idx)
        injury_codes_list = list()
        non_injury_codes_list = list()
        labels_list = list()
        for index,row in history_df.iterrows():
            code = row['healthCode']
            code_np = np.load(os.path.join(code_path, f'{code}.npy'))
            # print(code_np.shape)
            
            if row['health-history']=='Head Injury':
                injury_codes_list.append(code_np)
                labels_list.extend([0 for _ in range(code_np.shape[0])])
            else:
                non_injury_codes_list.append(code_np)
                labels_list.extend([1 for _ in range(code_np.shape[0])])
        total_injury__codes = np.vstack(injury_codes_list)
        total_labels = np.array(labels_list)
        print('total', total_codes.shape, total_labels.shape)
        isomap_analysis(total_codes, total_labels, cp)
```
